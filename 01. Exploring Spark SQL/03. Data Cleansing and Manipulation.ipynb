# Prerequisites

## Load Employee.scv into DataFrame

```python
df = spark.read.csv(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/employee.csv",
                    sep="|",
                    inferSchema=True,
                    header=True,
                    quote="'")
display(df)  
```

# Dropping Missing Records

```python
df.na.drop().display()
```

```python
df.na.drop(how="all").display()
```

```python
df.na.drop(subset=["name"]).display()
```

```python
df.na.drop(subset=["name","id"]).display()
```

```python
from pyspark.sql.functions import col
df.filter((col("id").isNotNull()) & (col("name").isNotNull())).display()
```

# Filling Missing Records

```python
df.na.fill(0).display()
```

```python
df.na.fill(0).na.fill("NULL IN SOURCE").display()
```

```python
df.na.fill({
    "id": -1,
    "name": "NA",
    "exp": 0,
    "gen": "NA",
    "dob": "1900-01-01",
    "company": "NULL IN SOURCE",
    "desig": "NULL IN SOURCE",
    "doj": "1900-01-01",
    "col_skills": "NULL IN SOURCE"
}).display()
```

```python
from pyspark.sql.functions import avg
mean = df.select(avg("exp")).collect()[0][0]
df.na.fill({
    "id":-1,
    "exp":int(mean),
    "gen":"NA"
}).display()
```
